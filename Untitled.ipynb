```python
Inflation_Rate['diff_1'] = Inflation_Rate.T10YIE - Inflation_Rate.T10YIE.shift()
Inflation_Rate.plot(y='diff_1')
plt.title('Inflation_Rate Time Series')
plt.show()
```

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(center=False,window=12).mean()
    rolstd = timeseries.rolling(center=False,window=12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(Inflation_Rate.T10YIE)
```

```python
test_stationarity(Inflation_Rate.diff_1[1:])
```

## Generate and interpret a PACF for the time series (and/or its differences).

**Generate and interpret a partial autocorrection function (PACF).**
PACF can help determine the AR parameter. Use stationary data for this.

```python
x_acf = pd.DataFrame(pacf(Inflation_Rate.T10YIE))
x_acf.plot(kind='bar')
plt.title("T10YIE PACF")
```

```python
x_acf = pd.DataFrame(pacf(Inflation_Rate.diff_1[1:]))
x_acf.plot(kind='bar')
plt.title("T10YIE Diff PACF")
plt.legend(['difference'])
```

## Generate five ARIMA specifications and evaluate their efficacy at modeling your data.

Use d=1 for all iterations to achieve stationarity. For p and q values, check increasingly up to 6 where we see spikes in the both ACF and PACF. For each iteration, print results summary and plot residuals as a scatter plot and histogram to evaluate for randomness and normal distribution.

```python
model010 = ARIMA(Inflation_Rate.diff_1[1:], order=(0,1,0))
model_fit_010 = model010.fit()
print(model_fit_010.summary())
print('Residuals Description')
print(model_fit_010.resid.describe())

residuals_010 = pd.DataFrame(model_fit_010.resid)
residuals_010.plot(legend=False)
plt.title('0, 1, 0 Time Series Residuals')

residuals_010.hist(bins=20)
plt.title('0, 1, 0 Residuals Histogram')

plt.show()
```

```python
model110 = ARIMA(Inflation_Rate.diff_1[1:], order=(1,1,0))
model_fit_110 = model110.fit()
print(model_fit_110.summary())
print('Residuals Description')
print(model_fit_110.resid.describe())

residuals_110 = pd.DataFrame(model_fit_110.resid)
residuals_110.plot(legend=False)
plt.title('1, 1, 0 Time Series Residuals')

residuals_110.hist(bins=20)
plt.title('1, 1, 0 Residuals Histogram')

plt.show()
```

```python
model111 = ARIMA(Inflation_Rate.diff_1[1:], order=(1,1,1))
model_fit_111 = model111.fit()
print(model_fit_111.summary())
print('Residuals Description')
print(model_fit_111.resid.describe())

residuals_111 = pd.DataFrame(model_fit_111.resid)
residuals_111.plot(legend=False)
plt.title('1, 1, 1 Time Series Residuals')

residuals_111.hist(bins=20)
plt.title('1, 1, 1 Residuals Histogram')

plt.show()
```

```python
model_011 = ARIMA(Inflation_Rate.diff_1[1:], order=(0,1,1))
model_fit_011 = model_011.fit()
print(model_fit_011.summary())
print('Residuals Description')
print(model_fit_011.resid.describe())

residuals_011 = pd.DataFrame(model_fit_011.resid)
residuals_011.plot(legend=False)
plt.title('0, 1, 1 Time Series Residuals')

residuals_011.hist(bins=20)
plt.title('0, 1, 1 Residuals Histogram')

plt.show()
```

```python

model_210 = ARIMA(Inflation_Rate.diff_1[1:], order=(2,1,0))
model_fit_210 = model_210.fit()
print(model_fit_210.summary())
print('Residuals Description')
print(model_fit_210.resid.describe())

residuals_210 = pd.DataFrame(model_fit_210.resid)
residuals_210.plot(legend=False)
plt.title('2, 1, 0 Time Series Residuals')

residuals_210.hist(bins=20)
plt.title('2, 1, 0 Residuals Histogram')

plt.show()


```

Performance Summary:

111 had an AIC = -534 and significant coefficient p-values. Residual errors had a mean of 0 and show a normal distribution.
